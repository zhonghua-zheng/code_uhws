```python
import xarray as xr
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import gc
```

# 2006

## Max

```python
# load CESM and get the date
df_cesm = pd.read_csv("/glade/scratch/zhonghua/urban_heat_train/data/002_2006_2015.csv")
df_cesm["date"] = df_cesm.time.str.slice(stop=10)
df_cesm_max = df_cesm[["lat","lon","date","TREFMXAV_U"]]
#df_cesm_min = df_cesm[["lat","lon","date","TREFMNAV_U"]]
```

```python
# load CMIP6 max and get the date
df_max = pd.read_csv("/glade/scratch/zhonghua/CMIP5-pred/urban_heat_CMIP_2006.csv")
df_max["date"] = df_max.time.str.slice(stop=10)

# merge based on lat, lon, and date, then save as csv
t0 = time.time()
df_max_final = pd.merge(df_max, df_cesm_max, how = "outer", on=["lat","lon","date"])
df_max_final_drop = df_max_final.drop(["time"],axis=1)
df_max_final_drop.to_csv("/glade/scratch/zhonghua/CMIP5-pred/urban_heat_full_2006.csv",index=False)
print(time.time()-t0)
```
